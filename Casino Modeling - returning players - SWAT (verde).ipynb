### Prep for Modeling

```python
s.loadactionset(actionset="sampling")

casino.srs(partInd=True,
    samppct=70,
    output={"casout":{"name":"casino_part", "replace":True},"copyVars":"all"})

casino_part = s.CASTable(name="casino_part")
```

```python
training   = casino_part.query('1 = _partind_')  
validation = casino_part.query('0 = _partind_') 
print("Training Rows: " + str(training.numrows()))
print("Validation Rows: " + str(validation.numrows()))
```

```python
target_str = "visit_2022"
target_lst= ["visit_2022"]

char_input = ["age","Brand_12m_ADW_group","c_country_cd","c_gender","c_tier_cd" ,"lifecycle_group","offer_type","Pref_game_type","region_code"]
num_input = ["Brand_12m_ADW","days_since_last_trip","num_trips_12M","num_trips_24M" ,"num_trips_ALL","patron_life","total_theo_12M","total_theo_24M" ,"total_theo_ALL"]


all= char_input + num_input + target_lst
```

### Build GB Model & Score Validation

```python
s.loadactionset('decisionTree')

trainingResultsGB = training.gbtreetrain(
        target=target_str,
        inputs=all,
        nominals=char_input + target_lst,
        casOut={"name":"gradboost_Py", "replace":True},
        seed=20,
        ntree=800,
        binOrder=True,
        varImp=True,
        nBins=20, 
        maxLevel=6, 
        maxBranch=2, 
        leafSize=5, 
        missing='USEINSEARCH', 
        minUseInSearch=1
    )

# make python object cas handle
gbtreeModel = s.CASTable("gradboost_Py")
```

```python
validation.gbtreeScore(
    modelTable=gbtreeModel, 
    copyVars=["_partind_", target_str],
    casOut={"name":"gradboost_Py_Scored", "replace":True},
    assessonerow=True
)
```

### Build score Random Forest Model

```python
s.loadactionset('decisionTree')

trainingResultsGB = training.foresttrain(
        target=target_str,
        inputs=all,
        nominals=char_input + target_lst,
        casOut={"name":"forest_Py", "replace":True},
        seed=20,
        ntree=800,
        binOrder=True,
        varImp=True,
        minUseInSearch=1
    )

# make python object cas handle
forestModel = s.CASTable("forest_Py")
```

```python
validation.forestScore(
    modelTable=forestModel, 
    copyVars=["_partind_", target_str],
    casOut={"name":"forest_Py_Scored", "replace":True},
    assessonerow=True
)
```

### Evaluate Model in Python

```python
 def prepROC(casScoreTabName,casAssessTabName,response_var,Yevent,Nevent,pvarname,pval):

    ctab = s.CASTable(casScoreTabName)             
    ctab.groupBy="_PartInd_"                         
    ctab.computedVars=["P1", "P0"]                   
    ctab.computedVarsProgram='''
       if strip(' ''' + Yevent + ''' ') eq strip(''' + pvarname + ''') then do;
         P1 = ''' + pval +''';
         P0 = 1 - ''' + pval + ''';
       end;
       else do;
         P1 = 1 - ''' + pval + ''';
         P0 = ''' + pval + ''';
       end;
    '''              
    ctab.percentile.assess(
        casOut={"name":casAssessTabName, "replace":True},
        nbins=10,
        cutStep = 0.01,
        inputs=['P1'],
        response=response_var, 
        event=Yevent, 
        pVar=['P0'], 
        pEvent=Nevent
    )
    return(ctab)
```

```python
gbtreeScored = prepROC("gradboost_Py_Scored","gradboost_Py_Assess",target_str,"1","0","_GBT_PredName_","_GBT_PredP_")
forestScored = prepROC("forest_Py_Scored","forest_Py_Assess",target_str,"1","0","_RF_PredName_","_RF_PredP_")
```
