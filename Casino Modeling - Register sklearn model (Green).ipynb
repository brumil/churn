### Import Pandas, Sklearn & SASctl packages

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
import xgboost as xgb

from sasctl import Session, register_model, publish_model, current_session
from sasctl.services import microanalytic_score as mas
from sasctl.services import model_repository as mr
from sasctl.services import model_management as mm
```

### Build sklearn model

```python
target= ["visit_2022"]

char_input = ["age","Brand_12m_ADW_group","c_country_cd","c_gender","c_tier_cd" ,"lifecycle_group","offer_type","Pref_game_type","region_code"]
num_input = ["Brand_12m_ADW","days_since_last_trip","num_trips_12M","num_trips_24M" ,"num_trips_ALL","patron_life","total_theo_12M","total_theo_24M" ,"total_theo_ALL"]

all_inputs=char_input + num_input
```

```python
train_x = raw_df[all_inputs]
train_y = raw_df[target]
```

```python
# Create an ML pipeline that imputes and one-hot encodes category columns then trains an xgboost model
cat_si_step = ('si', SimpleImputer(strategy='constant', fill_value='MISSING'))
cat_ohe_step = ('ohe', OneHotEncoder(handle_unknown='ignore'))
cat_steps = [cat_si_step, cat_ohe_step]
cat_pipe = Pipeline(cat_steps)
cat_transformers = [('cat', cat_pipe, char_input)]
ct = ColumnTransformer(transformers=cat_transformers)
ml_pipe = Pipeline([('transform', ct), ('xgb', xgb.XGBClassifier())])
ml_pipe.fit(train_x, train_y)
```

```python
ml_pipe.predict_proba(raw_df)
```

```python
ml_pipe.score(train_x,train_y)
```

### Register Model to Model Manager

```python
model_name = 'Python - OHE + xgBoost - v2'

# Register the model in Model Manager
register_model(ml_pipe,
    model_name,
    input=train_x,         # Use X to determine model inputs
    project='Casino Modeling - Returning Visits',  # Register in "Iris" project
    force=True)      # Create project if it doesn't exist    
```
